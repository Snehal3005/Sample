9. Which of the following three functions may be used to convert a value to an integer, a floating-point number, or a 
string?

    Function int() converts value to a integer value. e.g. function int(123.4) convert floating point number 123.4 to an

10. What is the error caused by this expression? What would you do about it?
'I have eaten ' + 99 + ' burritos.' 

    This expression gave a TypeError. It means only strings can be concatenated together. Integer and string cannot
    be concatenated directly. Integer need to be converted to string.

    Correction -
    
    'I have eaten ' + str(99) + ' burritos.'                                     
    
                    Or 

    'I have eaten ' + ‘99’ + ' burritos.'
