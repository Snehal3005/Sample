11. How can you tell the difference between break and continue?
    
    The Break statement terminate the current loop and resumes the execution from the next statement. 
    
    here is an example of break


```python
for i in range(0,10):
    print(i)   
    if i == 5:
        break;
```

The continue statement works opposite to break. It continues the execution of loop. When the continue statement is executed in the loop, the code inside the loop after the continue statement will be skipped and the next iteration of the loop will begin.

here is an example of continue

```python
for i in range(0,5):
    if i == 2:
        continue;
    print(i)    
```

13. Using a for loop, write a short programme that prints the numbers 1 to 10 Then, using a while loop, create an identical programme that prints the numbers 1 to 10.

```python
print('Numbers are:')
for i in range(1,11):
    print(i)
```

```python
print('Numbers are:')
i = 1
while i < 11:
    print(i)
    i += 1
```
