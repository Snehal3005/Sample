5. Create a programme that searches a folder tree for files with a certain file extension (such as .pdf or .jpg). Copy these files from whatever location they are in to a new folder.


```python
#search files with extension, copy all the files to a new folder

from functools import partial
from tkinter import *
from tkinter import messagebox
import os
import shutil
import docx

def search(list_box, file_name):
    if 'Copy_files' not in os.listdir():
        os.mkdir('Copy_files') #create a new folder in cwd
    
    listbox.delete(0,'end')
    search_file_type = file_name.get()
    file_path = []

    for folder , sub_folders , files in os.walk("c:\\Users\panka"):
        for file in files:
            if os.path.splitext(file)[1] == search_file_type: #compare extension of file in folder with search extension
                file_path.append(os.path.join(folder, file)) #append all files with path to file_path list
    
    i = 0
    for f in file_path: #file_path contains all files with given extension
        shutil.copy(f,'Copy_files') #copy files to new folder
        list_box.insert(i,f) #for displaying 
        i += 1
    messagebox.showinfo("Copy Files", "Files copied to 'Copy_files' folder successfully")


m = Tk()
file_name = StringVar()

label = Label(m,text="Enter the extension of file you would like to search....") 
entry = Entry(m,textvariable = file_name)
listbox = Listbox(m, height = 20, width = 100)
button = Button(m,text = 'Search', command = partial(search,listbox,file_name))
labelResult = Label(m,text = 'Search results are....')

label.grid(row = 0 , column = 0)
entry.grid(row = 1 , column = 0)
button.grid(row = 1 , column = 1)
labelResult.grid(row = 2, column = 0)
listbox.grid(row = 3 , column =0)

m.mainloop()
```
