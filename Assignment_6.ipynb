7. What are the values of the following expressions?

    'Hello'.upper()
    
    'Hello'.upper().isupper()
    
    'Hello'.upper().lower()

```python
#converts all characters of string to uppercase
print('Hello'.upper()) 

#converts all characters of string to uppercase, checks if string is all uppercase and if yes returns True, if no returns 
#False 
print('Hello'.upper().isupper()) 

#converts all characters of string to uppercase and then lowercase
print('Hello'.upper().lower())
```

8. What are the values of the following expressions?

    'Remember, remember, the fifth of July.'.split()

    '-'.join('There can only one.'.split())
    

```python
#split() method separates the words on space character and forms a list of words
print('Remember, remember, the fifth of July.'.split())

#First string is splitted into words using split() method
#Then each word is being joined with other with '.' using join() method
print('-'.join('There can only one.'.split()))
```
