4. If a PdfFileReader object’s PDF is encrypted with the password swordfish, what must you do before you can obtain Page objects from it?

    We use file.decrypt('swordfish') 
    
    explained in detail as follow:

```python
output_file = PyPDF2.PdfFileWriter()

num = pdfreader.numPages

for page_index in range(num):
    
    # Get the page at index idx
    page = pdfreader.getPage(page_index)
      
    # Add it to the output file
    output_file.addPage(page)

password = 'swordfish'

output_file.encrypt(password)

with open("my_encrypted_file.pdf", "wb") as file:
    
    # Write our encrypted PDF to this file
    output_file.write(file)
    
decrypted_file = PyPDF2.PdfFileReader(r'my_encrypted_file.pdf','rb')    

decrypted_file.decrypt(password)

print(decrypted_file.getPage(0).extractText())
```

12. What integers represent the levels of headings available in Word documents?

    0,1,2,3,4 represent the levels of headings available in Word documents


```python

# for question no. 10, 11, 12 combined code

doc1 = docx.Document() #create document object

doc1.add_heading('Sample', 0) #heading level 0
  
doc_para = doc1.add_paragraph('Word documents contain formatted text wrapped within three object levels. Lowest level- Run objects, Middle level- Paragraph objects and Highest level- Document object.So, we cannot work with these documents using normal text editors. But, we can manipulate these word documents in python using the python-docx module.')
  
doc_para.add_run('hey there, bold here').bold = True
doc_para.add_run(', and ')
doc_para.add_run('these words are italic').italic = True
  
doc1.add_page_break()
  
doc1.add_heading('Sample1 ', 2) #level 2
  
doc1.save('sample.docx')
```
